# **Preprocess the Tabular Data**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Load the tabular data
data = pd.read_csv("survey lung cancer.csv")  # Save the provided data as a CSV file

# Check for missing values
print("Missing values:\n", data.isnull().sum())

# Encode categorical variables
le_gender = LabelEncoder()
data['GENDER'] = le_gender.fit_transform(data['GENDER'])  # M -> 1, F -> 0

le_target = LabelEncoder()
data['LUNG_CANCER'] = le_target.fit_transform(data['LUNG_CANCER'])  # YES -> 1, NO -> 0

# Convert binary features (1, 2) to (0, 1)
binary_columns = [
    'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE',
    'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING',
    'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN'
]
for col in binary_columns:
    data[col] = data[col].map({1: 0, 2: 1})

# Scale numerical feature (AGE)
scaler = StandardScaler()
data['AGE'] = scaler.fit_transform(data[['AGE']])

# Separate features and target
X_tabular = data.drop('LUNG_CANCER', axis=1)
y_tabular = data['LUNG_CANCER']

# Split the tabular data into train, validation, and test sets (70-15-15 split)
X_tabular_temp, X_tabular_test, y_tabular_temp, y_tabular_test = train_test_split(
    X_tabular, y_tabular, test_size=0.15, random_state=42, stratify=y_tabular
)
X_tabular_train, X_tabular_valid, y_tabular_train, y_tabular_valid = train_test_split(
    X_tabular_temp, y_tabular_temp, test_size=0.1765, random_state=42, stratify=y_tabular_temp
)  # 0.1765 of 85% is ~15% of the total

print("Tabular Train Shape:", X_tabular_train.shape)
print("Tabular Validation Shape:", X_tabular_valid.shape)
print("Tabular Test Shape:", X_tabular_test.shape)
```

# **CT images**

```python
import os
import cv2
import numpy as np

def preprocess_image(filepath, target_size=(128, 128)):
    """Read and preprocess an image."""
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Failed to load image: {filepath}")
    img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
    img = img.astype("float32") / 255.0
    return img

def load_images_from_folder(folder, target_size=(128, 128)):
    """Load all images from a folder and its subfolders recursively."""
    images = []
    labels = []
    filenames = []
    
    print(f"Loading from folder: {folder}")
    for class_name in os.listdir(folder):
        class_path = os.path.join(folder, class_name)
        if not os.path.isdir(class_path):
            print(f"Skipping non-directory: {class_name}")
            continue
        # Map class to binary label
        print(f"Processing class: {class_name}")
        if "adenocarcinoma" in class_name.lower() or "large.cell.carcinoma" in class_name.lower() or \
           "malignant" in class_name.lower() or "squamous.cell.carcinoma" in class_name.lower():
            label = 1  # LUNG_CANCER = YES
            print(f"Mapping {class_name} to label 1")
        elif "bengin" in class_name.lower() or "normal" in class_name.lower():
            label = 0  # LUNG_CANCER = NO
            print(f"Mapping {class_name} to label 0")
        else:
            print(f"Unknown class: {class_name}")
            continue
        
        class_images = []
        class_filenames = []
        for root, _, files in os.walk(class_path):
            for filename in files:
                if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                    print(f"Skipping non-image file: {filename}")
                    continue
                filepath = os.path.join(root, filename)
                try:
                    img = preprocess_image(filepath, target_size)
                    class_images.append(img)
                    class_filenames.append(filename)
                except Exception as e:
                    print(f"Error processing {filepath}: {e}")
        
        print(f"Loaded {len(class_images)} images from class {class_name}")
        images.extend(class_images)
        labels.extend([label] * len(class_images))
        filenames.extend(class_filenames)
    
    images = np.array(images)
    labels = np.array(labels)
    print(f"Total images loaded from {folder}: {len(images)}")
    print(f"Unique labels in loaded data: {np.unique(labels, return_counts=True)}")
    return images, labels, filenames

# Define paths
dataset_dir = "LungcancerDataSet\Data"  # Update this to the correct path
train_dir = os.path.join(dataset_dir, "train")
valid_dir = os.path.join(dataset_dir, "valid")
test_dir = os.path.join(dataset_dir, "test")

# Load images
train_images, train_labels, train_filenames = load_images_from_folder(train_dir)
valid_images, valid_labels, valid_filenames = load_images_from_folder(valid_dir)
test_images, test_labels, test_filenames = load_images_from_folder(test_dir)
```

# **Pair the Tabular Data with Images**

```python
def sample_images(images, labels, target_count, target_label):
    """Sample images with a specific label, allowing replacement if necessary."""
    indices = np.where(labels == target_label)[0]
    if len(indices) == 0:
        raise ValueError(f"No images available with label {target_label}.")
    sampled_indices = np.random.choice(indices, target_count, replace=True)
    return images[sampled_indices], labels[sampled_indices]

def pair_data(X_tabular, y_tabular, images, labels):
    """Pair tabular data with images based on the target label."""
    pos_count = np.sum(y_tabular == 1)
    neg_count = np.sum(y_tabular == 0)
    print(f"Required positive samples (label 1): {pos_count}")
    print(f"Required negative samples (label 0): {neg_count}")
    
    pos_images, pos_labels = sample_images(images, labels, pos_count, 1)
    neg_images, neg_labels = sample_images(images, labels, neg_count, 0)
    
    paired_images = np.concatenate([pos_images, neg_images], axis=0)
    paired_labels = np.concatenate([pos_labels, neg_labels], axis=0)
    
    shuffle_indices = np.random.permutation(len(paired_labels))
    paired_images = paired_images[shuffle_indices]
    paired_labels = paired_labels[shuffle_indices]
    
    return paired_images, paired_labels

# Pair the data
train_images_paired, train_labels_paired = pair_data(X_tabular_train, y_tabular_train, train_images, train_labels)
valid_images_paired, valid_labels_paired = pair_data(X_tabular_valid, y_tabular_valid, valid_images, valid_labels)
test_images_paired, test_labels_paired = pair_data(X_tabular_test, y_tabular_test, test_images, test_labels)

# Add channel dimension for CNN
train_images_paired = train_images_paired[..., np.newaxis]
valid_images_paired = valid_images_paired[..., np.newaxis]
test_images_paired = test_images_paired[..., np.newaxis]

print("Paired Train Images Shape:", train_images_paired.shape)
print("Paired Validation Images Shape:", valid_images_paired.shape)
print("Paired Test Images Shape:", test_images_paired.shape)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten, Concatenate, Dropout
from tensorflow.keras.optimizers import Adam

# Image branch
image_input = Input(shape=(128, 128, 1), name="image_input")
x1 = Conv2D(32, (3, 3), activation="relu")(image_input)
x1 = MaxPooling2D((2, 2))(x1)
x1 = Conv2D(64, (3, 3), activation="relu")(x1)
x1 = MaxPooling2D((2, 2))(x1)
x1 = Flatten()(x1)
x1 = Dense(128, activation="relu")(x1)

# Tabular branch
tabular_input = Input(shape=(X_tabular_train_selected.shape[1],), name="tabular_input")
x2 = Dense(64, activation="relu")(tabular_input)
x2 = Dense(32, activation="relu")(x2)

# Combine branches
combined = Concatenate()([x1, x2])
x = Dense(64, activation="relu")(combined)
x = Dropout(0.3)(x)
output = Dense(1, activation="sigmoid")(x)

# Build and compile the model
model = Model(inputs=[image_input, tabular_input], outputs=output)
model.compile(optimizer=Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])

# Train the model
history = model.fit(
    [train_images_paired, X_tabular_train_selected], y_tabular_train,
    validation_data=([valid_images_paired, X_tabular_valid_selected], y_tabular_valid),
    epochs=10,
    batch_size=32,
    verbose=1
)

models['Deep Learning'] = model
```
