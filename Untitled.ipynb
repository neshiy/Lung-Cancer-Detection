```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the tabular dataset
df = pd.read_csv("survey lung cancer.csv")  # Update with your file path

# Rename columns to replace spaces with underscores
df.rename(columns={
    "CHRONIC DISEASE": "CHRONIC_DISEASE",
    "ALCOHOL CONSUMING": "ALCOHOL_CONSUMING",
    "SHORTNESS OF BREATH": "SHORTNESS_OF_BREATH",
    "SWALLOWING DIFFICULTY": "SWALLOWING_DIFFICULTY",
    "CHEST PAIN": "CHEST_PAIN"
}, inplace=True)

# Encode categorical variables
le_gender = LabelEncoder()
df["GENDER"] = le_gender.fit_transform(df["GENDER"])  # M/F to 0/1
le_lung_cancer = LabelEncoder()
df["LUNG_CANCER"] = le_lung_cancer.fit_transform(df["LUNG_CANCER"])  # YES/NO to 1/0

# Map binary features (1 = No, 2 = Yes) to 0/1
binary_cols = ["SMOKING", "YELLOW_FINGERS", "ANXIETY", "PEER_PRESSURE", "CHRONIC_DISEASE", 
               "FATIGUE ", "ALLERGY ", "WHEEZING", "ALCOHOL_CONSUMING", "COUGHING", 
               "SHORTNESS_OF_BREATH", "SWALLOWING_DIFFICULTY", "CHEST_PAIN"]
for col in binary_cols:
    df[col] = df[col].map({1: 0, 2: 1})

# Scale AGE
scaler = StandardScaler()
df["AGE"] = scaler.fit_transform(df["AGE"].values.reshape(-1, 1))

# Features and target
X_tabular = df.drop("LUNG_CANCER", axis=1)
y_tabular = df["LUNG_CANCER"]

print("Tabular Features Shape:", X_tabular.shape)
print("Tabular Labels Shape:", y_tabular.shape)
```

```python
from sklearn.model_selection import train_test_split

# First split: 70% train, 30% temp (validation + test)
X_tabular_train, X_tabular_temp, y_tabular_train, y_tabular_temp = train_test_split(
    X_tabular,
    y_tabular,
    test_size=0.3,
    stratify=y_tabular,
    random_state=42
)

# Second split: Split the temp set into 15% validation and 15% test
X_tabular_valid, X_tabular_test, y_tabular_valid, y_tabular_test = train_test_split(
    X_tabular_temp,
    y_tabular_temp,
    test_size=0.5,
    stratify=y_tabular_temp,
    random_state=42
)

print("Tabular Train Shape:", X_tabular_train.shape)
print("Tabular Validation Shape:", X_tabular_valid.shape)
print("Tabular Test Shape:", X_tabular_test.shape)
```

```python
import random

def sample_images(images, labels, target_count, target_label):
    """Sample images with a specific label."""
    indices = np.where(labels == target_label)[0]
    if len(indices) < target_count:
        print(f"Warning: Not enough images with label {target_label}. Using all available.")
        sampled_indices = indices
    else:
        sampled_indices = np.random.choice(indices, target_count, replace=False)
    return images[sampled_indices], labels[sampled_indices]

# Function to pair tabular data with images
def pair_data(X_tabular, y_tabular, images, labels):
    paired_images = []
    paired_labels = []
    
    # Count the number of positive and negative cases in the tabular data
    pos_count = np.sum(y_tabular == 1)
    neg_count = np.sum(y_tabular == 0)
    
    # Sample images for positive cases (LUNG_CANCER = YES)
    pos_images, pos_labels = sample_images(images, labels, pos_count, 1)
    # Sample images for negative cases (LUNG_CANCER = NO)
    neg_images, neg_labels = sample_images(images, labels, neg_count, 0)
    
    # Combine the sampled images
    paired_images = np.concatenate([pos_images, neg_images], axis=0)
    paired_labels = np.concatenate([pos_labels, neg_labels], axis=0)
    
    # Shuffle the paired data to match the tabular data order
    shuffle_indices = np.random.permutation(len(paired_labels))
    paired_images = paired_images[shuffle_indices]
    paired_labels = paired_labels[shuffle_indices]
    
    return paired_images, paired_labels

# Pair the data for each split
train_images_paired, train_labels_paired = pair_data(X_tabular_train, y_tabular_train, train_images, train_labels)
valid_images_paired, valid_labels_paired = pair_data(X_tabular_valid, y_tabular_valid, valid_images, valid_labels)
test_images_paired, test_labels_paired = pair_data(X_tabular_test, y_tabular_test, test_images, test_labels)

print("Paired Train Images Shape:", train_images_paired.shape)
print("Paired Validation Images Shape:", valid_images_paired.shape)
print("Paired Test Images Shape:", test_images_paired.shape)
```

```python
# Add channel dimension to images
train_images_paired = train_images_paired[..., np.newaxis]  # Shape: (216, 128, 128, 1)
valid_images_paired = valid_images_paired[..., np.newaxis]  # Shape: (46, 128, 128, 1)
test_images_paired = test_images_paired[..., np.newaxis]    # Shape: (47, 128, 128, 1)

# Convert tabular data to NumPy arrays
X_tabular_train = X_tabular_train.values
X_tabular_valid = X_tabular_valid.values
X_tabular_test = X_tabular_test.values
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate, Dropout

# Image branch (CNN)
image_input = Input(shape=(128, 128, 1), name="image_input")
x1 = Conv2D(32, (3, 3), activation="relu")(image_input)
x1 = MaxPooling2D((2, 2))(x1)
x1 = Conv2D(64, (3, 3), activation="relu")(x1)
x1 = MaxPooling2D((2, 2))(x1)
x1 = Flatten()(x1)
x1 = Dense(128, activation="relu")(x1)

# Tabular branch (Dense layers)
tabular_input = Input(shape=(15,), name="tabular_input")
x2 = Dense(64, activation="relu")(tabular_input)
x2 = Dense(32, activation="relu")(x2)

# Concatenate the two branches
combined = Concatenate()([x1, x2])
x = Dense(64, activation="relu")(combined)
x = Dropout(0.5)(x)
output = Dense(1, activation="sigmoid")(x)

# Define the model
model = Model(inputs=[image_input, tabular_input], outputs=output)

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Summary of the model
model.summary()
```

```python
# Train the model
history = model.fit(
    [train_images_paired, X_tabular_train], y_tabular_train,
    validation_data=([valid_images_paired, X_tabular_valid], y_tabular_valid),
    epochs=10,
    batch_size=32
)

# Evaluate on the test set
test_loss, test_accuracy = model.evaluate([test_images_paired, X_tabular_test], y_tabular_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```
